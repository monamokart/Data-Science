```python
def plot_cm(y_test, y_pred):  # to plot a confusion matrix
    m = confusion_matrix(y_test, y_pred, labels=[0, 1])
    m = m.astype('float') / m.sum(axis=1)[:, np.newaxis]

    df_cm = pd.DataFrame(m, columns=["Ant", "Grapes"], index=["Ant", "Grapes"])
    df_cm.index.name = 'Actual'
    df_cm.columns.name = 'Predicted'
    sn.heatmap(df_cm, annot=True, cmap="Reds")
```

## II.1 - Normalize and shuffle

The objective here is to predict whether a drawing is an ant or a grape. The data is described by the value of the pixels between 0 and 255. We should normalize them to avoid that some pixels have more importance than others. Moreover the data are not shuffled, let's change that:

```python
X_train = np.concatenate((ant,grapes), axis=0)
y_train = np.concatenate((np.zeros(1000,), np.ones(1000,)), axis=0).astype("int64")  # create the y_train

permut_1 = np.random.permutation(np.arange(2000))  # shuffle the data
X_train = X_train[permut_1]
y_train = y_train[permut_1]

permut_2 = np.random.permutation(np.arange(1000))
X_test = test_images[permut_2]
y_test = test_labels[permut_2]

scaler = MinMaxScaler()  # normalize the data
scaler.fit(X_train)
X_train_n = scaler.transform(X_train)
X_test_n = scaler.transform(X_test)
```

## II.2 - Reduction (PCA)

Here the main problem is that the number of features is very large so the computation will be very long, maybe we could see if a PCA reduction is a solution. Let's see with the perceptron algorithm (the simplest)

```python
variances = [0.8, 0.9, 0.95, 0.99, 0.995]
n_component = []
test_scores_pca = []
train_scores_pca = []

for var in variances :
    pca = PCA(var)
    pca.fit(X_train_n)
    train_pca = pca.transform(X_train_n)
    test_pca = pca.transform(X_test_n)
    perc = Perceptron()
    perc.fit(train_pca, y_train)
    
    train_scores_pca.append(perc.score(train_pca, y_train))
    test_scores_pca.append(perc.score(test_pca, y_test))
    n_component.append(train_pca.shape[1])

perc = Perceptron()
perc.fit(X_train_n, y_train)
full_train_score = perc.score(X_train_n, y_train)
full_test_score = perc.score(X_test_n, y_test)

    
fig, axes = plt.subplots(ncols=2, figsize=(15, 5))
ax = axes.ravel()
ax[0].plot(variances, train_scores_pca, label="Training scores", color ='orangered')
ax[0].plot(variances, test_scores_pca, label="Test scores", color = 'mediumblue')
ax[0].axhline(full_train_score, color='orangered', ls="--")
ax[0].axhline(full_test_score, color='mediumblue', ls='--')
ax[0].set_title("PCA\nScores with respect to total variance")
ax[0].legend()
ax[1].bar(np.array(variances).astype(str), n_component, color='lightseagreen')
ax[1].set_title("Number of component \nwith respect to total variance")
plt.show()
```

So we can see that the solution is not obvious, the importance is to determine the best test score, and the graph shows that in some case of PCA it is better to reduce the dimension of the data. First we will choose the best classifier and its parameters and then we will see how we need to reduce dimension with this classifier.

Here we use a pre-trained model and we change the input and output layers to be adapted to our data. We will train just on the new layers and we won't touch the pre-trained weights in the other layers.

```python
X_train_cnn2 = np.zeros((X_train_n.shape[0], 28, 28, 3))  # adapt the shape of the data for the pre-trained CNN
X_test_cnn2 = np.zeros((X_test_n.shape[0], 28, 28, 3))

X_train_cnn3 = np.zeros((X_train_n.shape[0], 32, 32, 3))
X_test_cnn3 = np.zeros((X_test_n.shape[0], 32, 32, 3))

X_train_cnn2[:,:,:,0] = X_train_cnn[:,:,:,0]
X_train_cnn2[:,:,:,1] = X_train_cnn[:,:,:,0]
X_train_cnn2[:,:,:,2] = X_train_cnn[:,:,:,0]

X_test_cnn2[:,:,:,0] = X_test_cnn[:,:,:,0]
X_test_cnn2[:,:,:,1] = X_test_cnn[:,:,:,0]
X_test_cnn2[:,:,:,2] = X_test_cnn[:,:,:,0]

for i in range(X_train_cnn3.shape[0]):
    X_train_cnn3[i] = resize(X_train_cnn2[i], (32,32))
for i in range(X_test_cnn3.shape[0]):
    X_test_cnn3[i] = resize(X_test_cnn2[i], (32,32))
```

```python
vgg16 = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))  # VGG16
for layer in vgg16.layers[:]:  #  we freeze the pre-trained weights
    layer.trainable = False
model_cnn2 = Sequential()
model_cnn2.add(vgg16)
model_cnn2.add(Flatten())
model_cnn2.add(Dense(512, activation='relu'))
model_cnn2.add(Dense(1, activation='sigmoid'))

model_cnn2.compile(loss='binary_crossentropy', optimizer=optimizers.Adam(lr=1e-4),metrics=["accuracy"])
```

```python
tf.config.run_functions_eagerly(True)

history = model_cnn2.fit(X_train_cnn3, y_train, validation_split=0.2, epochs=10, batch_size=batch_size, verbose=2)
dt = time()-t0

plt.plot(history.history['accuracy'], color="orangered")
plt.plot(history.history['val_accuracy'], color="mediumblue")
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

It seems that the test scores are still good with PCA, therefore we could use it so that the calculation time becomes shorter.

## V.2 - Simple classification

As we saw earlier that at a large scale we could see that one class had more white pixels than another, we can take a combination of these pixels to make some features.
Therefore if we want to do a simple and easy classifier with only 2 features, we can use the means and the standard deviations of the images, it could be enough

```python
X_train_mean = X_train[:,500:].mean(axis=1)
X_test_mean = X_test[:,500:].mean(axis=1)

X_train_std = X_train[:,500:].std(axis=1)
X_test_std = X_test[:,500:].std(axis=1)

X_train_ = np.concatenate((X_train_mean[:,np.newaxis], X_train_std[:,np.newaxis]), axis=1)
X_test_ = np.concatenate((X_test_mean[:,np.newaxis], X_test_std[:,np.newaxis]), axis=1)
```

```python
svm_mean = SVC(kernel='rbf')
svm_mean.fit(X_train_, y_train)
pred_mean = svm_mean.predict(X_test_)
train_score_mean = svm_mean.score(X_train_, y_train)
test_score_mean = svm_mean.score(X_test_, y_test)
scores["2features_svm"] = [train_score_mean, test_score_mean]
predictions["2features_svm"] = pred_mean

print("Training score : ", train_score_mean)
print("Test score :", test_score_mean)
print("Confusion matrix")
plot_cm(y_test, pred_mean)
```

Here with 2 features we can see that we have a better accuracy than the perceptron with 784 features, it shows that it is easy to have a sufficient classifier with few features but to reach a very good accuracy we need very complex models such as CNN.
