# Importing Keras

````python
# Importing the Keras main module forcing tensorflow 1.x backend
import tensorflow as tf
import keras
print("Using tensorflow version " + str(tf.__version__))
print("Using keras version " + str(keras.__version__))
````

## Loading and preparing the MNIST dataset

Again, load the dataset via ```keras.datasets```, turn train and test labels into one-hot encoding, and reshape and normalize data as in the first exercise. 

````python
from keras.datasets import mnist
# START CODE HERE
(train_images, y_train),(test_images, y_test) = mnist.load_data()
print("Size of training set :", train_images.shape)
print("Size of testing set :", test_images.shape)
n_train = train_images.shape[0]
n_test = test_images.shape[0]

# END CODE HERE
````

````python
from keras.utils.np_utils import to_categorical
# START CODE HERE
train_labels = to_categorical(y_train, 10)
test_labels = to_categorical(y_test, 10)
# END CODE HERE
````

````python
from sklearn.preprocessing import StandardScaler
# Reshape to proper images with 1 color channel according to backend scheme
img_rows, img_cols = train_images.shape[1], train_images.shape[2]

# START CODE HERE
train_images = train_images.reshape((n_train, img_rows*img_cols))
test_images = test_images.reshape((n_test, img_rows*img_cols))
# END CODE HERE

# Cast pixels from uint8 to float32
train_images = train_images.astype('float32')

# Now let us normalize the images so that they have zero mean and standard deviation
# Hint: are real testing data statistics known at training time ?
# START CODE HERE
scaler = StandardScaler()
scaler.fit(train_images)
train_images = scaler.transform(train_images)
test_images = scaler.transform(test_images)
# END CODE HERE
````

## Defining the neural network architecture (i.e., the network model)

You can take a look at this [cheatsheet](https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Keras_Cheat_Sheet_Python.pdf) for some basic commands to use keras.

First, try to replicate the classifier of the first exercise. Secondly, create a fully connected network.
For the fully connected layer, you can for example use this architecture: 
$$ (784) \rightarrow (300) \rightarrow (10) $$
For this first implementation of the network, use only sigmoid activations in the hidden layer. Remember to use the right output activation function ! 

````python
# The Sequential module is a container for more complex NN elements and
# defines a loop-less NN architecture
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten

# START CODE HERE
input_shape = 784
middle_shape = 300
output_shape = 10

#model 0
model0=Sequential()
model0.add((Dense(output_shape,activation='softmax',input_dim=input_shape)))

#fully connected model
model = Sequential()
model.add(Dense(middle_shape,activation='sigmoid',input_dim=input_shape))
model.add(Dense(output_shape,activation='softmax',input_dim=middle_shape))

# END CODE HERE
````

Instantiate a SGD optimizer with a tentative learning rate of $\\eta = 10^{-2}$ and, using the appropriate loss function (which is called, in keras, ```'categorical_crossentropy'```) and compile the model.

````python
# The optimizers module provides a number of optimization algorithms for updating
# a netwok parameters accoridng to the computed error gradints
from keras import optimizers

# START CODE HERE
eta = 0.01
opt = optimizers.SGD(learning_rate=eta)
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
model0.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])

# END CODE HERE
# We can now have a look at the defined model topology
model.summary()
````

````python
#model0
model0.summary()
````

## Training the network

Train the model for 10 epochs using the ```.fit()``` method, validating the model at each epoch and keeping track of the training history for later plotting. Make sure you enable ```.fit()``` verbose mode in order to visualize the training.

In order to accelerate training, use the ```batch_size``` option of ```.fit()```, which will process a batch of examples at the same time, and make one update for all of them, averaged over the gradients for each training example of the batch. You can begin with a small size, and experiment with a larger size later.

````python
# This is where the actual training-testing happens
# Number of epochs we want to train
epochs = 10

# START CODE HERE
history=model.fit(train_images, train_labels, epochs=epochs, verbose=2, batch_size=20, validation_data=(test_images,test_labels))
# END CODE HERE
````

## Visualizing the network performance

Visualize the training history using the ```pyplot``` package:
- In one graph, plot the train and vaidation loss functions,
- In another graph, the train and validation accuracy.
By comparing the training the testing curves, what can we conclude about the quality of the training ?

````python
# We now want to plot the train and validation loss functions and accuracy curves
import matplotlib.pyplot as plt
print(history.history.keys())

# summarize history for loss
plt.figure(figsize=(15,7))
plt.plot([i for i in range(1,11)],history.history['loss'],color='b',label='Train loss function')
plt.plot([i for i in range(1,11)],history.history['val_loss'],color='g',label='Validation loss function')
plt.title('Train and validation loss functions')
plt.legend()
plt.show()

# summarize history for accuracy
plt.figure(figsize=(15,7))
plt.plot([i for i in range(1,11)],history.history['accuracy'],color='b',label='Train accuracy')
plt.plot([i for i in range(1,11)],history.history['val_accuracy'],color='g',label='Validation accuracy')
plt.title('Train and validation accuracy')
plt.legend()
plt.show()
````

It appears clearly that as the epoch number grows, the training and test data get similar accuracy and loss functions. Therefore, for a good number of epochs, the training is well realised. 

## Experiments

Note down the performance of the larger network in terms of training and validation accuracy as a reference (save the loss/accuracy graphs of the network).

Then, experiment as follow and compare performance with the reference scenario:

*  Experiment increasing the size of the batch and compare the performance with reference.
*  Experiment replacing the sigmoid activations with Relus and note what happens.

$$ (784) \rightarrow (300) \rightarrow (128) \rightarrow (84) \rightarrow (10) $$

To realise these experiments, we reuse the code from above and modify the parameters of our model.  

**1. Experiment increasing the size of the batch**

````python
model = Sequential()
model.add(Dense(middle_shape, activation='sigmoid', input_dim=input_shape))
model.add(Dense(output_shape, activation='softmax', input_dim=middle_shape))
opt = optimizers.SGD(learning_rate=0.01)
model.compile(loss='categorical_crossentropy',optimizer=opt,metrics=['accuracy'])
history2=model.fit(train_images, train_labels, epochs=epochs, verbose=2, batch_size=40, validation_data=(test_images,test_labels))
````

````python
# We now want to plot the train and validation loss functions and accuracy curves

# summarize history for loss
plt.figure(figsize=(15,7))
plt.plot([i for i in range(1,11)],history.history['loss'],color='b',label='Train loss function - Initial model')
plt.plot([i for i in range(1,11)],history.history['val_loss'],color='g',label='Validation loss function - Initial model')
plt.plot([i for i in range(1,11)],history2.history['loss'],color='b', ls="--", label='Train loss function - New model')
plt.plot([i for i in range(1,11)],history2.history['val_loss'],color='g', ls="--", label='Validation loss function - New model')
plt.title('Train and validation loss functions for initial and new model')
plt.legend()
plt.show()

# summarize history for accuracy
plt.figure(figsize=(15,7))
plt.figure(figsize=(15,7))
plt.plot([i for i in range(1,11)],history.history['accuracy'],color='b',label='Train loss function - Initial model')
plt.plot([i for i in range(1,11)],history.history['val_accuracy'],color='g',label='Validation loss function - Initial model')
plt.plot([i for i in range(1,11)],history2.history['accuracy'],color='b', ls="--", label='Train accuracy - New model')
plt.plot([i for i in range(1,11)],history2.history['val_accuracy'],color='g', ls="--", label='Validation accuracy - New model')
plt.title('Train and validation accuracy')
plt.legend()
plt.show()
````

It appears that the loss is higher and the accuracy lower for a bigger batch. Therefore, it is more intresting for our case to use a lower batch. 

**2. Experiment replacing the sigmoid activations with Relus**

````python
model3 = Sequential()
model3.add(Dense(middle_shape,activation='relu',input_dim=input_shape))
model3.add(Dense(output_shape,activation='softmax',input_dim=middle_shape))
opt = optimizers.SGD(learning_rate=eta)
model3.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
history3=model3.fit(train_images, train_labels, epochs=epochs, verbose=2, batch_size=20, validation_data=(test_images,test_labels))
````

````python
# We now want to plot the train and validation loss functions and accuracy curves

# summarize history for loss
plt.figure(figsize=(15,7))
plt.plot([i for i in range(1,11)],history.history['loss'],color='b',label='Train loss function')
plt.plot([i for i in range(1,11)],history.history['val_loss'],color='g',label='Validation loss function')
plt.plot([i for i in range(1,11)],history3.history['loss'],color='b', ls="-.", label='Train loss function - New model')
plt.plot([i for i in range(1,11)],history3.history['val_loss'],color='g', ls="-.", label='Validation loss function - New model')
plt.title('Train and validation loss functions')
plt.legend()
plt.show()

# summarize history for accuracy
plt.figure(figsize=(15,7))
plt.plot([i for i in range(1,11)],history.history['accuracy'],color='b',label='Train accuracy')
plt.plot([i for i in range(1,11)],history.history['val_accuracy'],color='g',label='Validation accuracy')
plt.plot([i for i in range(1,11)],history3.history['accuracy'],color='b', ls="-.", label='Train accuracy - New model')
plt.plot([i for i in range(1,11)],history3.history['val_accuracy'],color='g', ls="-.", label='Validation accuracy - New model')
plt.title('Train and validation accuracy')
plt.legend()
plt.show()
````

It appears clearly that the accuracy is a lot lower than on the initial model, and respectively the loss is higher. The accuracy is totally dropping for high epochs. Using relu instead of sigmoid for the activation function is therefore not necessary for our data. 

**3. Experiment with a larger architecture**

````python
model4 = Sequential()
model4.add(Dense(300,activation='sigmoid',input_dim=input_shape))
model4.add(Dense(128,activation='sigmoid',input_dim=300))
model4.add(Dense(84,activation='sigmoid',input_dim=128))
model4.add(Dense(output_shape, activation='softmax',input_dim=84))
opt = optimizers.SGD(learning_rate=eta)
model4.compile(loss='categorical_crossentropy',optimizer=opt,metrics=['accuracy'])
history4=model4.fit(train_images, train_labels, epochs=epochs, batch_size=20,verbose=2,validation_data=(test_images,test_labels))
````

````python
# We now want to plot the train and validation loss functions and accuracy curves

# summarize history for loss
plt.figure(figsize=(15,7))
plt.plot([i for i in range(10)],history.history['loss'],color='b',label='Train loss function - Initial model')
plt.plot([i for i in range(10)],history.history['val_loss'],color='g',label='Validation loss function - Initial model ')
plt.plot([i for i in range(1,11)],history4.history['loss'],color='b', ls=":", label='Train loss function - New model')
plt.plot([i for i in range(1,11)],history4.history['val_loss'],color='g', ls=":", label='Validation loss function - New model')
plt.title('Train and validation loss functions')
plt.legend()
plt.show()

# summarize history for accuracy
plt.figure(figsize=(15,7))
plt.plot([i for i in range(10)],history.history['accuracy'],color='b',label='Train accuracy - Initial model')
plt.plot([i for i in range(10)],history.history['val_accuracy'],color='g',label='Validation accuracy - Initial model')
plt.plot([i for i in range(1,11)],history4.history['accuracy'],color='b', ls=":", label='Train accuracy - New model')
plt.plot([i for i in range(1,11)],history4.history['val_accuracy'],color='g', ls=":", label='Validation accuracy - New model')
plt.title('Train and validation accuracy')
plt.legend()
plt.show()
````

Those curves prove that having a more complex and larger structure is not necessary as the accuracy is the same at the end and the loss functions also. Using 2 layers is therefore enough.
