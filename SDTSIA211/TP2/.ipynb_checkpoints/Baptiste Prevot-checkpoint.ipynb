```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


def load_breastcancer(filename):
    """
    Cette fonction lit le fichier filename, par exemple
    filename = 'wdbc_M1_B0.data'
    Elle retourne 
    X : une matrice de caracteristiques
    y : un vecteur des classes tel que si y[i] = 1, la tumeur est maligne
        et si y[i] = -1, la tumeur est benigne

    Pour plus d'infos sur la base de donnees,
    https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Prognostic%29
    """

    data = np.loadtxt(filename, delimiter=',')

    # la colonne 0 ne nous interesse pas ici
    y = data[:, 1] * 2 - 1
    X = data[:, 2:]

    # Standardisation de la matrice
    X = X - np.mean(X, axis=0)
    X = X / np.std(X, axis=0)

    return X, y

```

```python
X,y = load_breastcancer('wdbc_M1_B0.data')
```

# Question 2.3

We consider $(v,a)$ as a vector of size m+1 <br>
$N(v,a) = \frac{1}{2}\sum\limits_{i=1}^m v_i^2$

We then choose H : 
$H(Z) = \sum\limits_{j=1}^n max(0,1-z_i) = \sum\limits_{j=1}^n h(z_i)$ <br><br>
diag(y) is a square matrix of size n, where the diagonal values are the values of y. The rest is 0.
X_1 is a matrix of size 569*31 built with X concatenated to a column of ones.

Finally we set M = diag(y)*X_1

We now have :
$f(v,a) = N(v,a) + cH(M(v,a))$

$N$ and $H$ are classic convex functions<br>
$M$ is obviously a linear operator<br>

as $0 \in dom(H)$ and $0 \in dom(N)$, $0 \in relint(dom(H) - Mdom(N))$. <br>
We apply 2.4.2 :<br><br>
$\partial f(v,a) = \partial N(v,a) +cM^T\partial H(M(v,a))$


We already show that H is clearly separable.

$\partial H(v) = \partial h (v)^n$, with the same $\partial h (v)$ as before.


$\partial N(v,a) = (v,0)$ It is a vector built with v, and an added 0 at the end.

```python
M = np.diag(y)@np.concatenate((X,np.ones((len(X),1))), axis = 1)
```

```python
def subgrad_H(Z): #returns one of the subgradient of the function in z
    ret = []
    for i in range(len(Z)):
        if Z[i]<=1:
            ret.append(-1)
        else:
            ret.append(0)
    
    return ret
    
def f(v,a):
    
    sum1 = np.linalg.norm(v)**2
    sum2 = H(M@np.append(v,a))
    
    return 0.5*sum1 + sum2


def subgrad_f(v,a):
    
    Z = M.dot(np.append(v,a))
    
    subg_H = subgrad_H(Z)
    
    return np.append(v,0) + M.T@subg_H
```

# Question 2.5

```python
def subgradient_descent(tolerance,max_iter):
    
    x = np.zeros(len(X[0])+1)
    k = 0
    subg = subgrad_f(x[:-1],x[-1])
    
    while np.linalg.norm(subg)>tolerance and k<max_iter:
        gamma = 1/(k+1)
        x = x-gamma*subg
        
        subg = subgrad_f(x[:-1],x[-1])
        k+=1
        print(np.linalg.norm(subg))

    return x[:-1],x[-1]
```

# Question 4.2

If $x + \rho^{-1}\phi \geq 0$, $g(x,\phi) = -\frac{1}{2\rho}\phi^2 + \frac{\rho}{2}(x+\rho^{-1}\phi)^2$ and $\nabla_xg(x,\phi) = \rho(x + \frac{\phi}{2})$, $\nabla_\phi g(x,\phi) = x$ <br>
If $x + \rho^{-1}\phi = 0$, $\nabla_xg(x,\phi) = 0$ and $\nabla_\phi g(x,\phi) = x = -\frac{\phi}{\rho}$

If $x + \rho^{-1}\phi \leq 0$, $g(x,\phi) = -\frac{1}{2\rho}\phi^2 $ and $\nabla_xg(x,\phi) = 0$, $\nabla_\phi g(x,\phi) = -\frac{\phi}{\rho}$ <br><br>

In conclusion, $\nabla_xg(x,\phi) = max(0,\rho(x + \frac{\phi}{2}))$, $\nabla_\phi g(x,\phi) = max(x,-\frac{\phi}{\rho})$

# Question 4.3

If $x + \rho^{-1}\phi \geq 0$, $\nabla_xg(x,\phi) = \rho(x + \frac{\phi}{2})$ and $\nabla_\phi g(x,\phi) = x$ <br>
$\nabla_x^2g(x,\phi) = \rho > 0$, $\nabla_\phi^2 g(x,\phi) = 0$ <br><br>

If $x + \rho^{-1}\phi \leq 0$, $\nabla_xg(x,\phi) = 0$, $\nabla_\phi g(x,\phi) = -\frac{\phi}{\rho}$ <br>
$\nabla_x^2g(x,\phi) = 0$, $\nabla_\phi^2 g(x,\phi) = -\frac{1}{\rho} < 0$ <br><br>

In conclusion, the restriction of $g$ on $x$ is convexe, and the restriction of $g$ on $\phi$ is concave.

```python
def grad_g_x(x,phi,rho):
    return rho*max(0, x + (phi/rho))

def grad_G1_x(x,phi,rho): #here x is a vector
    ret = []
    for i in range(len(x)):
        ret.append(grad_g_x(x[i],phi[i],rho))
    
    return ret

def grad_G2_x(x,psi,rho): #here x is a vector
    ret = []
    for i in range(len(x)):
        ret.append(grad_g_x(1+x[i],psi[i],rho))
    
    return ret


def grad_g_phi(x,phi,rho):
    return max(-phi/rho, x)

#Later, x stands for (v,a,ξ)

def grad_L_x(v, a, xi, phi, psi, rho): #Calculates the gradient for a given (phi,psi)
    
    x = np.concatenate((np.concatenate((v,[a])),xi))
    
    #we use the separability of the 2 sums
    #for a given ϕ and a given ψ, each sum is a separable function of a linear combination of (v,a,ξ)
    #I'll use the proposition 2.4.2 again
    
    M1 = np.concatenate((np.zeros((len(X),len(X[0])+1)),-1*np.identity(len(X))), axis=1)
    
    M = np.diag(y)@np.concatenate((X,np.ones((len(X),1))), axis=1)
    M2 = np.concatenate((M,-1*np.identity(len(X))), axis=1)
    
    
    grad0 = np.zeros(len(x))
    for i in range(len(v)):
        grad0[i] = v[i]
    
    grad1 = (M1.T)@grad_G1_x(M1@x,phi,rho)
    
    grad2 = (M2.T)@grad_G2_x(M2@x,psi,rho)

    return grad0 + grad1 + grad2
```

```python
def lagrangian_gradient_descent(phi,psi,tolerance,max_iter):
    
    x = np.zeros(len(X[0])+1+len(X))
    k = 0
    subg = grad_L_x(x[:len(X[0])], x[len(X[0])], x[len(X[0])+1:], phi, psi, 2)
    
    while np.linalg.norm(subg) > tolerance and k < max_iter:
        gamma = 1/(k+1)
        x = x-gamma*subg
        
        subg = grad_L_x(x[:len(X[0])], x[len(X[0])], x[len(X[0])+1:], phi, psi, 2)
        k+=1
        print(np.linalg.norm(subg))

    return x[:len(X[0])], x[len(X[0])], x[len(X[0])+1:]
```
