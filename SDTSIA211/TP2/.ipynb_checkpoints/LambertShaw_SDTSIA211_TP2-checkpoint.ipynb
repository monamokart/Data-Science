```python
#Importation
import numpy as np
```

```python
def load_breastcancer(filename):
    """
    Cette fonction lit le fichier filename, par exemple
    filename = 'wdbc_M1_B0.data'
    Elle retourne 
    X : une matrice de caracteristiques
    y : un vecteur des classes tel que si y[i] = 1, la tumeur est maligne
        et si y[i] = -1, la tumeur est benigne

    Pour plus d'infos sur la base de donnees,
    https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Prognostic%29
    """

    data = np.loadtxt(filename, delimiter=',')

    # la colonne 0 ne nous interesse pas ici
    y = data[:, 1] * 2 - 1
    X = data[:, 2:]

    # Standardisation de la matrice
    X = X - np.mean(X, axis=0)
    X = X / np.std(X, axis=0)

    return X, y
```

# **2. Méthode du sous-gradient**


Question 2.1 :
On cherche à minimiser la quantité $\displaystyle \min_{v\in\mathbb{R}^m,a\in\mathbb{R},\xi \in \mathbb{R}^n}\frac{1}{2}\displaystyle\sum_{j=1}^{m}v_j^2+c\sum_{i=1}^{n}\xi_i$


Or, on sait que $\xi_i$ a des contraintes, il doit être à la fois positif et plus grand que $1-y_i(x_i^Tv+a)$ pour $i$ fixé. C'est à dire qu'il doit être plus grand que le maximum entre 0 et $1-y_i(x_i^Tv+a)$. Or, vu que l'on cherche le minimum de la quantité initiale et que les $\xi_i$ sont ponctués d'un signe positif, il suffit de les prendre égaux au maximum, car il n'existe pas de nombre plus petit satisfaisant les contraintes.

On peut ainsi réecrire le problème :
$\displaystyle \min_{v\in\mathbb{R}^m,a\in\mathbb{R},\xi \in \mathbb{R}^n}\frac{1}{2}\displaystyle\sum_{j=1}^{m}v_j^2+c\sum_{i=1}^{n}max(0,1-y_i(x_i^Tv+a))$

et si $(v,a,\xi)$ est optimum de (1), $(v,a)$ est optimum de (2).

Réciproquement, si $(v,a)$ est l'optimum du problème (2), on peut contruire l'optimum du problème (1) comme étant $(v,a,\xi)$ avec $\xi_i=max(0,1-y_i(x_i^Tv+a))$.

Question 2.3 :

On pose : $N(v,a)=\displaystyle \sum_{j=1}^mn_j(v_j)$ avec $n_j(v_j)=\frac{1}{2}v_j^2$ d'ou $N(v,a)$ est séparable.

Puis : $M(v,a)=y(x^Tv+a)$ qui est une fonction linéaire.

Et enfin : $H(x)=\displaystyle\sum_{i=0}^nh_i(x_i)$ avec $h_i=max(0,1-x_i)$

Les fonctions N et H sont des sommes de fonctions convexes, elles sont donc convexes. De plus, M est linéaire. D'après la question précédente, H est différentiable sauf en 1, comme somme de fonction de la forme $h(z)=max(1-z)$.

On peut  utiliser la propriété 2.4.2 du cours (avec égalité puisque dom H = $\mathbb{R}^n$ et conclure :

$\partial f(v,a) = \partial N(v,a)+cM^T\partial H(M(v,a))$

On calcule ensuite les quantités suivantes :

$\partial N(v,a)= v_1 \times v_2 \times ... \times v_n \times \partial N(a)= (v,0)$

$\partial H(m) =  \partial h_1(m_1) \times \partial h_2 (m_2)\times ... \times \partial h_n(m_n) $
avec les $h_i$ égaux à la fonction $h$ de la question précédente  (donc leur sous-gradient aussi) et $m_i=M(v,a)[i]$.

Question 2.4 :


```python
def f(v,a):
    #Computing f(v,a)
    c=1
    X, y = load_breastcancer('sample_data/wdbc_M1_B0.data.txt')
    N=  np.sum(v**2)/2
    H=0
    for i in range(len(X)):
        H+= max(0,1-y[i]*(np.dot(X[i],v)+a))      
    f=N+c*H
    return f
    
def df(v,a):
    #Computing one of the subgradient of f
    c=1
    X, y = load_breastcancer('sample_data/wdbc_M1_B0.data.txt')
    dH=np.zeros(len(X))
    for i in range(len(dH)):
        if y[i]*(np.dot(X[i],v)+a)<1:
            dH[i]=-1
        else:
            dH[i]=0      
    Xe=np.insert(X,30,1,axis=1)
    M=Xe.T *y
   
    v2=np.append(v,0)
    df= v2 + c*np.dot(M,dH)  #because subgrad(N)=(v,0)
    return df

print(f(np.ones(30),1))
print(df(np.ones(30),1))
```

Question 2.5 :

```python
def subgradient(v0,a0,end=500) :
  '''
    Use the subgradient method on the function of the problem
    v0 and a0 are the initial values for v and a
    end is the number of iteration we want to make
    Return the final value for v and a, and the value of the function f at that time
  '''
  ak = a0
  vk = v0
  dfk=df(v0,a0)
  k = 0
  while k < end :
    k+=1
    dfk = df(vk,ak)
    vk = vk + 1/(k+1)*dfk[:30]
    ak = ak + 1/(k+1)*dfk[-1]
  return (vk,ak,f(vk,ak))

subgradient(np.ones(30),0)
```

# **3 Méthode du sous gradient stochastique**

Question 3.1 :

$I$ suit une loi uniforme sur $\{1,2,....,n\}$, la probabilité $\mathbb{P}(I=i)=\frac{1}{n}$. On calcule donc l'espérance, c'est à dire le membre de droite :

$\mathbb{E}[f_I(v,a)]=\displaystyle \sum_{i=1}^{n}\frac{1}{n}f_i(v,a)$

$\mathbb{E}[f_I(v,a)]=\displaystyle\frac{1}{n}\sum_{i=1}^{n}(\frac{1}{2}\sum^{m}_{j=1}v_j^2 + cn\max(0,1-y_i(xi^Tv+a)))$

$\mathbb{E}[f_I(v,a)]=\displaystyle\frac{1}{2}\sum_{j=1}^{m}v_j^2+c\sum_{i=1}^{n}\max(0,1-y_i(xi^Tv+a)))$

Finalement on obtient bien : $\mathbb{E}[f_I(v,a)]=f(v,a)$

Question 3.2 :

On réutilise la décomposition vu à la question 2.3 en remplaçant H par la fonction de la question 2.2 (et plus la fonction séparable décrite à ce moment là), de même on pose $M_i(v,a)=y_i(xi^Tv+a)$ :

$\partial f_i(v,a) = \partial N(v,a)+cnM_i^T\partial H(M_i(v,a))$

$\partial f_i(v,a) = (v,0)+cnM_i^T\partial H(M_i(v,a))$

avec $\partial H(M_i(v,a))= \left \{
   \begin{array}{r c l}
      \{-1\}  &si& m_j<1 \\
      [-1,0]  &si& m_j=1 \\
      \{0\}  &si& m_j>1 \\
   \end{array}
   \right .$
   
avec $m_j$ la $j^{ème}$ composante de $M_i(v,a)$ et on fait ça pour toutes les composantes.

Question 3.3 :

```python
def dfi(v,a,i):
  #Computing one subgradient of fi
  c=1
  X, y = load_breastcancer('sample_data/wdbc_M1_B0.data.txt')
  dhM = np.zeros(len(X[0])+1)
  if y[i]*(np.dot(X[i],v)+a)<1:
      dhM[i] = -1 * y[i]*X[i]
  else:
      dhM=0      
  
  v2 = np.append(v,0)
  df= v2 + len(X)*c*dhM #because subgrad(N)=(v,0)
  return df

def stochgradient(v0,a0,end=500):
  n = len(X)
  ak = a0
  vk = v0
  k=0
  while k < end :
    k+=1
    i = np.random.randint(n)
    df_i = dfi(vk,ak,i)
    vk = vk + 1/(k+1)*df_i[:30]
    ak = ak + 1/(k+1)*df_i[-1]
  return(vk, ak)

res = stochgradient(np.ones(30),0,2)
print(res)
print(f(res[0],res[1]))
```

Question 4.2:

SI $x_i+\frac{\phi_i}{\rho}<0$ (ie.$x_i<-\frac{\phi_i}{\rho}$) alors  $g_i(x,\phi)=-\frac{1}{2\rho}\phi_i^2$ donc $\nabla_xg_i(x,\phi)=0$ et $\nabla_\phi g_i(x,\phi)=-\frac{\phi_i}{\rho}$.

Sinon $g_i(x,\phi)=-\frac{1}{2\rho}\phi_i^2+\frac{\rho}{2}(x_i+\rho^{-1}\phi_i)^2$ donc  $\nabla_xg_i(x,\phi)=\rho(x_i+\rho^{-1}\phi_i)$ et $\nabla_\phi g_i(x,\phi)=-\frac{\phi_i}{\rho}+\frac{\rho}{2}\frac{2}{\rho}(x_i+\frac{\phi_i}{\rho})=x_i$, d'où le résultat, où $max$ est le maximum terme à terme.



Question 4.3:

La fonction $x\rightarrow g(x,\phi)$ est la composition de la fonction convexe $u\rightarrow \frac{\rho}{2}u^2$ avec la fonction également convexe $x\rightarrow max(0, x+\rho^{-1}\phi)$ (plus un terme constant),donc celle-ci est convexe.

D'après la question précédente, la dérivée seconde de $\phi\rightarrow g(x,\phi)$ est négative quelle que soit la valeur de $\phi$, donc la fonction $\phi\rightarrow g(x,\phi)$ est concave.

```python
rho=2 

def g(x,phi):
  return -1/(2*rho)*phi**2+ rho*(np.maximum(0,x+phi/rho))**2/2

def grad_x_g(x, phi):
    return rho * np.maximum(0, x + phi/rho)

def grad_phi_g(x, phi):
    return np.maximum(-phi/rho, x)

def grad_v_L(v,a,xi,phi,psi):
  tmp = -xi+1-y*(np.dot(X,v)+a)
  return v-np.matmul(y*X.T,grad_x_g(tmp,psi))

def grad_a_L(v,a,xi,phi,psi):
  tmp = -xi+1-y*(np.dot(X,v)+a)
  return -np.sum(y*grad_x_g(tmp,psi))

def grad_xi_L(v,a,xi,phi,psi):
  tmp = -xi+1-y*(np.dot(X,v)+a)
  return 1-(grad_x_g(-xi,phi)+grad_x_g(tmp,psi))

def L(v,a,xi,phi,psi):
  tmp1= 0.5*np.sum(v**2)
  tmp2= np.sum(xi)
  tmp3=np.sum(g(-xi,phi))
  tmp4= np.sum(g(-xi+1-y*(np.dot(X,v)+a),psi))
  return tmp1+tmp2+tmp3+tmp4

def line_search(v,a,xi,phi,psi, gammak) :
  '''
    Using the Taylor-based line search in order to find the best gamma for the gradient method.
  '''
  b = 2*gammak
  c = 0.5
  l=0
  bal=b*c**l
  L_xk = L(v,a,xi,phi,psi)
  dv, da, dxi= grad_v_L(v,a,xi,phi,psi), grad_a_L(v,a,xi,phi,psi),grad_xi_L(v,a,xi,phi,psi)
  left_exp=L(v-bal*dv,a-bal*da,xi-bal*dxi,phi,psi)
  right_exp = L_xk+np.dot(dv,v-bal*dv)+np.dot(da,a-bal*da)+np.dot(dxi,xi-bal*dxi)+1/(2*bal)*(np.linalg.norm(dv*bal)**2+np.linalg.norm(da*bal)**2+np.linalg.norm(dxi*bal)**2)

  while left_exp > right_exp :
    l+=1
    bal=b*c**l
    left_exp=L(v-bal*dv,a-bal*da,xi-bal*dxi,phi,psi)
    right_exp =L_xk+np.dot(dv,v-bal*dv)+np.dot(da,a-bal*da)+np.dot(dxi,xi-bal*dxi)+1/(2*bal)*(np.linalg.norm(dv*bal)**2+np.linalg.norm(da*bal)**2+np.linalg.norm(dxi*bal)**2)

  return(bal)

        
def min_L(phik,psik):
  v = np.zeros(30)
  a = 0
  xi = np.zeros(569)
  gamma=0.0015
  tmp = -xi+1-y*(np.dot(X,v)+a)
  dv= grad_v_L(v,a,xi,phik,psik)
  
  da= grad_a_L(v,a,xi,phik,psik)
  dxi= grad_xi_L(v,a,xi,phik,psik)
 
  iter=0
  while(np.sum(dv**2)+np.sum(dxi**2)+da**2)>1:
     
    v-=gamma*dv
    a-=gamma*da
    xi-=gamma*dxi
    dv= grad_v_L(v,a,xi,phik,psik)
    da= grad_a_L(v,a,xi,phik,psik)
    dxi= grad_xi_L(v,a,xi,phik,psik)
    #gamma= line_search(v,a,xi,phik,psik, gamma)     Le line search ne fonctionne pas bien 
    iter+=1
  print(iter)
  print(L(v,a,xi,phik,psik)) 
  return v,a,xi

phi0=np.zeros(569)
min_L(phi0,phi0)
```

La méthode du Lagrangien augmenté s'écrit donc:

```python
def Aug_Lagrangian_method():
  iter_max=200
  phik=phi0
  psik=phi0
  for i in range(iter_max):
    v,a,xi= min_L(phik,psik)
    phik+= rho*grad_phi_L(v,a,xi,phik,psik)
    psik+= rho*grad_psi_L(v,a,xi,phik,psik)
  return v,a,xi

Aug_Lagrangian_method()
```

# **5 Comparaison**

Le premier algorithme, la méthode du sous gradient, est assez long à tourner, et la condition d'arrêt n'est pas très précise parce qu'elle ne repose que sur un nombre d'itération et aucune assurance de convergence.

Par définition, la méthode du gradient stochastique peut paraître moins efficace car elle ne prend qu'une seule des composantes en compte lors de la descente de gradient. Dans notre cas, il semblerait qu'elle ne soit pas efficace du tout, sachant que le résultat trouvé est encore très haut pour une minimisation, mais nous soupçonnons des erreurs dans ce code.

Enfin, la méthode du lagrangien augmentée est beaucoup plus complexe et prends un gros temps à tourner, elle semble cependant donné des résultats plus justes !
